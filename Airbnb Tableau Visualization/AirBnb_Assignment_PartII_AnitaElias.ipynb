### Task 2 -> Part I - 'Price'
---

```python
#Write your code here for plotting the distribution of price
import matplotlib.pyplot as plt
bin_start = 0
bin_end = 2000
bin_width = 50
air = sns.histplot(air_df['price'], alpha=0.7, color='blue',bins=range(bin_start, bin_end, bin_width))
plt.axvline(air_df.price.mean(), color ='red')
plt.axvline(air_df.price.median(), color ='yellow')
plt.title('Distribution of Booking Prices')
plt.xlabel('Price in Euros')
plt.ylabel('Frequency')
plt.show()
#Also printing skew value:
print(air_df.price.skew())
```

**Observation**:

### Here are the inferences based on observations:
1) There is a steep drop-off in frequency as prices rise, highlighting the concentration of listings in the more affordable range. 
2) As the price for rooms increase, the number of bookings made comes down. The least bookings are for rooms that cost above 500 Euros. This pattern suggests that luxury accommodations are much less common.Â 
3) The maximum number of bookings are being made for the rooms in the more affordable range of 50 - 150 Euros. 
4) The distribution is a right skewed distribution. 

### Task 2 -> Part II - 'Distribution of Reviews'
---

```python
##Write your code here for plotting the distribution of reviews
import matplotlib.pyplot as plt
bin_start = 0
bin_end = 2000
bin_width = 50
air = sns.histplot(air_df['number_of_reviews'], alpha=0.7, color='blue',bins=range(bin_start, bin_end, bin_width))
plt.title('Distribution of Reviews is lower for majority of Listings')
plt.xlabel('Distribution of Reviews')
plt.ylabel('Frequency')
plt.axvline(air_df.number_of_reviews.mean(), color ='red')
plt.axvline(air_df.number_of_reviews.median(), color ='yellow')
plt.show()
#Also printing skew value:
print(air_df.number_of_reviews.skew())
```

**Observation**:

### Here are the inferences based on observations:
1) A large majority of listings have a relatively low number of reviews. 
2) The first bin appears to contain the highest frequency with around 4500 listings, which is approximately 70% of all listings combined - 6173. 
3) The frequency decreases significantly as the number of reviews increases. This suggests that newer or less popular listings dominate the platform, or that guests tend to leave reviews infrequently. 
4) The distribution is a right skewed distribution. 

### Task 2 -> Part III - 'Review of Score Ratings'
---

```python
##Write your code here for plotting the distribution of ratings
import matplotlib.pyplot as plt

air = sns.histplot(air_df['review_scores_rating'], alpha=0.7, color='blue',bins=10)
plt.title('Histogram for Distribution of Review of Score Ratings')
plt.xlabel('Distribution of Review of Score Ratings')
plt.ylabel('Frequency')
plt.axvline(air_df.review_scores_rating.mean(), color ='red')
plt.axvline(air_df.review_scores_rating.median(), color ='yellow')
plt.show()
#Also printing skew value:
print(air_df.review_scores_rating.skew())
```

**Observation**:

1) The histogram for 'review_scores_rating' shows a left-skewed distribution, indicating that most properties have high ratings. 
2) The majority of the properties received a rating in the highest bin range of **4.5 to 5.0 star rating**. This could be indicative of a positive customer experience, positive brand perception or a bias towards leaving positive reviews.
3) There is a significant drop in frequency as the rating decreases, with the next most populated bin being **4.0 to 4.5 stars** containing only roughly around 400 or so properties.

# Task 4
---

You want to observe the relationship between the reviews(given by `number_of_reviews`) and the ratings received (`review_scores_rating`) by different AirBnB listings.

For this, plot both a scatterplot and a jointplot. What can you say about the relationship between ratings and reviews?

```python
##Write your code here for scatterplot
sns.scatterplot(data=air_df,
                x="number_of_reviews",
                y="review_scores_rating",
                hue="room_type")
plt.title('Scatterplot depicting relationship between Reviews and Review Score Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('Review Score Ratings')
plt.show()
```

```python
##Write your code here for jointplot
sns.jointplot(data=air_df, x="number_of_reviews", y="review_scores_rating", hue="room_type")
plt.xlabel('Number of Reviews')
plt.ylabel('Review Score Ratings')
plt.show()
```

**Observation** - 

1) From the scatterplot, it appears that there is a positive relationship between the number of reviews and the review scores ratings. This means that as the number of reviews increases, the review scores ratings also tend to increase.

2) In the jointplot, we see that the distribution of review scores ratings becomes more concentrated at higher values as the number of reviews increases.

3) Overall, we can say that there is a positive correlation between the number of reviews and the review scores ratings, indicating that listings with more reviews tend to have higher ratings.

2) After 2010, we see a generally upward trend in the average review scores ratings, indicating a positive reception and possibly an improvement in the quality of listings and services offered.
3) However, after 2016, there is a downward trend in the average review scores ratings. This could be due to a variety of factors, such as an increase in the number of listings leading to more competition, changes in the types of listings being offered, or shifts in customer preferences or external environmental factors like onset of Covid.

You wish to identify the relationship between the various ratings each of the listing has received from the customers.
These ratings have been summarized below:

```python
review_columns = ['review_scores_rating',
 'review_scores_accuracy',
 'review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']
print(review_columns)
```

### Plot a heatmap of the correlation matrix of the above ratings variables and document your observations. You can use this link to understand these variables further - https://www.airbnb.co.in/help/article/1257/star-ratings

```python
sns.heatmap(air_df[review_columns].corr(), cmap="Greens", annot=True)
plt.show()
```

**Observation** - 

1) Based on the heatmap, it appears that there is a generally high positive correlation between the different review scores. This means that listings with high ratings in one category tend to have high ratings in others.

2) This insight is valuable as it indicates that customer ratings across different aspects of the listing (accuracy, cleanliness, check-in experience, communication, location, and value) are positively related. This suggests that customers who are satisfied with one aspect of their stay are likely to be satisfied with other aspects as well.

3) This information can be used by AirBnB hosts and management to understand the interconnectedness of customer satisfaction across different dimensions and to focus on improving overall quality and customer experience.
